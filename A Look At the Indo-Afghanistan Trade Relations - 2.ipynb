### 5. Analysing India's Imports from Afghanistan

```python
# Importing all the required tools we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
```

```python
# Importing the data
import_data = pd.read_excel("Afghanistan data1.xlsx", sheet_name = "Sheet11", index_col = 0)
import_data.to_csv("Afghanistan data1.csv", encoding = 'utf-8', index = False)
import_data
```

```python
# Reindexing Columns and Rows
import_data.columns = import_data.iloc[2]
import_data = import_data.iloc[3:,].reindex()
import_data
```

```python
import_data.set_index(import_data.iloc[:, 1].values, inplace = True)
import_data.drop(["HSCode", "Commodity"], axis = 1, inplace = True)
import_data
```

```python
# Naming the columns and rows
import_data.rename_axis(index= "Commodity", columns = "Years", inplace = True)
import_data
```

```python
# Separating the table in two parts based on value and quantity
value_imports = import_data.iloc[:, :9]
quantity_imports = import_data.iloc[:, 9:]
value_imports
```

```python
quantity_imports
```

```python
# Removing rows with all Nan values
value_imports.dropna(how = "all", inplace = True)
value_imports
```

```python
quantity_imports.dropna(how = "all", inplace = True)
quantity_imports
```

```python
len(value_imports)
```

```python
len(quantity_imports)
```

```python
# Filling the Nan values with 0's
value_imports.fillna(0, inplace = True)
value_imports
```

```python
quantity_imports.fillna(0, inplace = True)
quantity_imports
```

```python
# Describing the Value of Imports
value_imports.info()
```

```python
value_imports.describe()
```

Here we can see that the average imports have seen a steady rise since 2020-21 when Taliban took control of Afghanistan. The highest average imports were seen in the year 2023-24 of about \\$16.05 million. 

```python
# Average Import value
value_imports.T.mean()
```

```python
# Describing the data of quantity of Imports from Afghanistan
quantity_imports.info()
```

```python
quantity_imports.describe()
```

```python
# Average Import Quantity
quantity_imports.T.mean()
```

```python
# Plotting the graph of the table in parts
value_imports_1 = value_imports.iloc[:20].T
value_imports_2 = value_imports.iloc[20:].T
```

```python
value_imports_1
```

```python
value_imports_2
```

```python
# Plotting the line plots as subplots
fig, (ax0, ax1) = plt.subplots(nrows = 2, 
                               ncols = 1, 
                               figsize = (30, 35))

ax0.plot(value_imports_1.index,
        value_imports_1);
ax1.plot(value_imports_2.index,
        value_imports_2)

ax0.set(title = "Plotting the Value of Imports to Afghanistan for the first 20 commodities",
        xlabel = "Years",
        ylabel = "Value of Imports in Millions ($)")
ax0.legend(value_imports_1, loc = "upper left", bbox_to_anchor=(1, 1));

ax1.set(title = "Plotting the Value of Imports to Afghanistan for the second 20 commodities",
        xlabel = "Years",
        ylabel = "Value of Imports in Millions ($)")
ax1.legend(value_imports_2, loc = "upper left", bbox_to_anchor=(1, 1));

fig.suptitle('India-Afghanistan Commodity-wise Imports Analysis', 
             fontsize=30, 
             fontweight='bold');
```

Here we can see that products like edible vegetables (highest being \\$51 million in 2023-24), fruits (highest being \\$380 million in 2023-24), oil seeds and olea (\\$130 million in 2019-20), and tea and coffee (highest being \\$61 million in 2023-24) are some of the major imports of India from Afghanistan.

```python
# Plotting the graph of the table in parts
quantity_imports_1 = quantity_imports.iloc[:20].T
quantity_imports_2 = quantity_imports.iloc[20:].T
```

```python
quantity_imports_1
```

```python
quantity_imports_2
```

```python
# Plotting the line plots as subplots
fig, (ax0, ax1) = plt.subplots(nrows = 2, 
                               ncols = 1, 
                               figsize = (30, 35))

ax0.plot(quantity_imports_1.index,
        quantity_imports_1);
ax1.plot(quantity_imports_2.index,
        quantity_imports_2)

ax0.set(title = "Plotting the Quantity of Imports to Afghanistan for the first 20 commodities",
        xlabel = "Years",
        ylabel = "Quantity of Imports in Millions ($)")
ax0.legend(quantity_imports_1, loc = "upper left", bbox_to_anchor=(1, 1));

ax1.set(title = "Plotting the Quantity of Imports to Afghanistan for the second 20 commodities",
        xlabel = "Years",
        ylabel = "Quantity of Imports in Millions ($)")
ax1.legend(quantity_imports_2, loc = "upper left", bbox_to_anchor=(1, 1));

fig.suptitle('India-Afghanistan Commodity-wise Imports Analysis', 
             fontsize=30, 
             fontweight='bold');
```

Here we can see the major imports from Afghanistan in terms of the quantity of items are: edible fruits (highest being 116 thousand units), edible vegetables (highest at 48 thousand units), wool hair and woven fabric (highest at 249 thousand units in 2019-20) and oil seeds and olea fruits (highest at 10 thousand million at 2020-21).

```python
# Now removing the commodities with 0 values
value_imports = value_imports.replace(0, np.nan)
value_imports = value_imports.dropna(how = "any", axis = 0)
value_imports
```

```python
len(value_imports)
```

```python
# Plotting the remaining commodity graph
plt.figure(figsize = (15, 10))
plt.plot(value_imports.T.index, value_imports.T.iloc[:]);
plt.xlabel("Years")
plt.ylabel("Value of Imports")
plt.title("Value of Imports to Afghanistan")
plt.grid()
plt.legend(value_imports.T.columns, loc = "upper left", bbox_to_anchor=(1, 1))
plt.show();
```

Here we can see the major import commodities are edible fruits, edible vegetables, lac, gums and resins and coffee and tea etc. 

```python
# Similarly Plotting the Quantity of the commodities without 0 values
quantity_imports = quantity_imports.replace(0, np.nan)
quantity_imports = quantity_imports.dropna(how = "any", axis = 0)
quantity_imports
```

```python
len(quantity_imports)
```

```python
# Plotting the remaining commodity graph
plt.figure(figsize = (15, 10))
plt.plot(quantity_imports.T.index, quantity_imports.T.iloc[:]);
plt.xlabel("Years")
plt.ylabel("Quantity of Imports")
plt.title("Quantity of Imports to Afghanistan")
plt.grid()
plt.legend(quantity_imports.T.columns, loc = "upper left", bbox_to_anchor=(1, 1))
plt.show();
```

Here too we see the top imports are edible fruits, vegetables and oil seeds and olea fruits etc.
