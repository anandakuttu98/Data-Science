### 6. Analysing the Total Trade Balance between India and Afghanistan

```python
# Importing all the required tools we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
```

```python
# Importing the data
trade_data = pd.read_excel("Afghanistan data1.xlsx", sheet_name = "Sheet12", index_col = 0, header = 1)
trade_data.to_csv("Afghanistan data1.csv", encoding = 'utf-8', index = False)
trade_data
```

```python
# Renaming the columns 
trade_data.rename(columns = {"India's Exports to Afghanistan in Millions US$" : "Exports to Afghanistan (in Millions US $)",
                            "India's Imports from Afghanistan in Million US$": "Imports from Afghanistan (in Millions US $)",
                            "Total Trade in Millions US$": "Total Trade (in Millions US $)",
                            "Net Exports (Exports - Imports) in Million US$": "Net Exports (Exports - Imports) (in Million US$)"},
                 inplace = True)
trade_data
```

```python
# Describing the trade data
trade_data.describe()
```

Here we can see that the average exports to Afghanistan over the past 2 and a half decades have been worth \\$352 million, with the minimum trade conducted in 1998-99 for about \\$12 million and the maximum trade happening in 2019-20 about \\$997 million. 

Similarly, the mean imports have been \\$203 and the minimum trade conducted in 1996-97 worth \\$3 million and the maximum trade taking place in 2023-24 for about \\$642 million. 

The overall trade between both countries have been around \\$555 million and the net exports i.e. Exports minus Imports have on an average been \\$149 million, which is net positive for Indian exports.

```python
# Plotting the trade data separately for each column i.e. Exports, Imports, Total trade and Net Exports
trade_data_export = trade_data.iloc[:, 0:1]
trade_data_import = trade_data.iloc[:, 1:2]
trade_data_total = trade_data.iloc[:, 2:3]
trade_data_netexp = trade_data.iloc[:, 3:4]
```

```python
# Now Ploting the data in subplots
fig, (ax0, ax1, ax2, ax3) = plt.subplots(nrows = 4, 
                                         ncols = 1, 
                                         figsize = (30, 35))

ax0.plot(trade_data.index,
        trade_data_export);
ax1.plot(trade_data.index,
        trade_data_import)
ax2.plot(trade_data.index,
        trade_data_total)
ax3.plot(trade_data.index,
        trade_data_netexp);

ax0.set(title = "Plotting the trend of Export to Afghanistan over the years",
        xlabel = "Years",
        ylabel = "Value of Export in Millions ($)")
ax0.legend(trade_data.columns[0], loc = "upper left", bbox_to_anchor=(1, 1));

ax1.set(title = "Plotting the trend of Imports to Afghanistan over the years",
        xlabel = "Years",
        ylabel = "Value of Imports in Millions ($)")
ax1.legend(trade_data.columns[1], loc = "upper left", bbox_to_anchor=(1, 1));

ax2.set(title = "Plotting the trend of Total Trade to and from Afghanistan over the years",
        xlabel = "Years",
        ylabel = "Value of Total Trade in Millions ($)")
ax2.legend(trade_data.columns[2], loc = "upper left", bbox_to_anchor=(1, 1));

ax3.set(title = "Plotting the trend of Net Export to Afghanistan over the years",
        xlabel = "Years",
        ylabel = "Value of Net Export in Millions ($)")
ax3.legend(trade_data.columns[3], loc = "upper left", bbox_to_anchor=(1, 1));

fig.suptitle('India-Afghanistan Trade Analysis', 
             fontsize=30, 
             fontweight='bold');
```

The above graphs show trends in exports, imports, total trade and net exports over the years. 

1. The first plot shows the trejectory of exports to Afghanistan over the years, wherein the exports kept steadily rising over the years until 2019-20 when it reached its peak at \\$997 million worth of trade in goods and services but has since kept falling year on year. The main reason that could be attributed is the takeover of power by Taliban regime and hence dismissal of all diplomatic and economic ties with Afghanistan by the International Community including India. The other reason could be the spread of covid-19 pandemic which disrupted much of the supply chains across the world and hence the affected exports. 

> If we go back a little earlier in time, in the late 1990s the exports were rather low due to poor ties with Afghanistan and considering India's economy was just emerging from a huge debt crisis and socio-economic and political challenges. There were also quite an amount of hostility between two countries and the Taliban regime was not recognised internationally even then. On an average the exports were around \\$23.37 worth of goods and serives.

> However, the relationship slowly picked up over the years from 2002-03 onwards, when NATO forces invaded Afghanistan and formed a new government led by Hamid Karzai. The average exports rose to about \\$247 million between 2002-03 and 2010-11 and further increased to an average of \\$470 million by 2014-15. When Ashraf Ghani took over the reigns of Afghanistan in 2014-15 the exports steadily rose to an average of \\$713 million until 2020-21, following which there has been a steady fall seen in total exports averaging around \\$448 million. 

3. Similarly, when it comes to imports, the average imports from Afghanistan to India was about \\$17.84 million in the late 1990s and early 2000s with the lowest amount of imports registered in 1996-97 at \\$3.05 million. However, the imports began improving since 2003-04, averaging around \\$45 million until 2006-07, and the suceeding years saw a huge rise in imports to about \\$144 million on average in value with an mean growth rate of 41.79% year on year till 2013-14. The imports saw a further increase to about \\$410 million from 2014-15 to 2021-22. Though the imports fell a little in 2022-23 amounting to about \\$452 million, it regained its upward trend in 2023-24 increasing by 41% amounting to \\$642 million.

4. The third graph shows the total trade i.e. Exports plus Imports of goods and services over the years. Between 1996-97 and 2002-03, the mean trade between both countries was just \\$46 million, which improved over the years from 2003-04 to 2010-11 averaging around \\$356 million and further increased to \\$712 million until 2016-17. The total trade crossed 1 billion dollars mark in the year 2017-18 and averaged around \\$1.2 billion until 2021-22. Subsequent years in 2022-23 saw a slight dip in total trade by 16.47%.

5. India has largely been a net exporter to Afghanistan barring a few years. However, in recent years 2022-23 and 2023-24, India has been importing more goods and services from Afghanistan than exporting, accounting for \\$15 million in 2022-23 and as high as \\$286 million in 2023-24. The last time net exports were negative was back in 1998-99 amounting to \\$15.33 million and in 2000-01 accounting to about \\$73 hundred thousand. However, the net exports of India rose year on year making it \\$338 million in 2009-10 and further rising to \\$467 million in 2019-20, which was the highest amount of net exports from India to Afghanistan witnessed so far. Although, net exports have been fluctuating over the years, sometimes rising and then falling, the average net exports between 2002-03 to 2014-15 has been about \\$215 million and between 2015-16 and 2023-24 has been about \\$ 167 million. 


```python
# Calculating Growth Rate of all the columns
trade_data["Growth Rate of Exports (%)"] = trade_data_export.pct_change()*100
trade_data["Growth Rate of Imports (%)"] = trade_data_import.pct_change()*100
trade_data["Growth Rate of Total Trade (%)"] = trade_data_total.pct_change()*100
trade_data["Growth Rate of Net Exports (%)"] = trade_data_netexp.pct_change()*100
trade_data
```

```python
# Filling the Nan value with 0
trade_data.fillna(0, inplace = True)
trade_data
```

```python
# Assigning the growth rates to a variable called growth data
growth_data = trade_data.iloc[:, 4:]
growth_data
```

```python
# Plottiing the graph of all the growth rates
plt.figure(figsize = (40, 20))
plt.plot(trade_data.index, growth_data);
plt.xlabel("Years")
plt.ylabel("Growth Rates")
plt.title("Plotting Growth Rates in Trade with Afghanistan")
plt.grid()
plt.legend(growth_data.columns, loc = "upper left", bbox_to_anchor=(1, 1))
plt.show();
```

Here we can witness the growth rates are hovering around +- 20 percentage for all the variables, however the most fluctuation is seen in net exports which dipped as low as -1003% in 2001-02 and rose up by 1720% in 2023-24. 

Growth of exports has been negative in the recent past reflecting the fall in exports year on year since 2020-21 and growth of imports have been unsteady althroughout, however has seen a positive trend last year by 41%. Though total trade between both countries saw a negative trend from 2020-21, it has bounced back to a positive figure of 12% in 2023-24.
